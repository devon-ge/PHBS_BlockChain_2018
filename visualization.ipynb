# Implementation of course project for Blockchain

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

## transactions per block

```python
data = pd.read_csv(r'data\transactions-per-block.csv', encoding='utf-8')
```

```python
data[:651]
```

```python
data['dates']=pd.to_datetime(data['date'])
```

```python
plt.plot(data['dates'], data['volume'])

plt.xlabel('Date')
plt.ylabel('Transactions per block')

plt.title("Bitcoin Transactions per block")
plt.savefig(r'images\transactions_per_block.png')
plt.show()
```
